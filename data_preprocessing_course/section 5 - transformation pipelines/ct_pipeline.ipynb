# **Example: Combining column transformer and piplines**
- ### Apply simple imputer with strategy median to all numerical variables.
- ### Apply simple imputer with strategy most frequent to all categorical variables
- ### Apply one hot encoding to categorical variables

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer, make_column_selector
from get_path import get_path as gp
```

```python
df = pd.read_csv(gp())
```

```python
transformer = ColumnTransformer([
    # numerical cleaning transformer

    ("numerical", 
     SimpleImputer(strategy = "median"), 
        make_column_selector(dtype_exclude="object")),
        
     #categorical cleaning+encoding pipeline

    ("categorical", 
     Pipeline([
         ("clean", SimpleImputer(strategy = "most_frequent")), 
         ("encode", OneHotEncoder())]), 
     make_column_selector(dtype_include="object"))
])
```

```python
transformer.fit_transform(df)
```
