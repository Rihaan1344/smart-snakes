# **Exercise 1**
- Load **sample_dataset.csv**
- Replace missing categorical values with 'N'
- Replace missing numerical values with mean value

```python
cleaner = ColumnTransformer([
    ("categorical_transformer", SimpleImputer(strategy="constant", fill_value="N"), make_column_selector(dtype_include="object")),
    ("numerical_transformer", SimpleImputer(strategy="mean"), make_column_selector(dtype_exclude="object"))
])
```

```python
cleaned = cleaner.fit_transform(df)
cleaned
```

```python
cleaned[0]
```

```python
cleaned = pd.DataFrame(
    cleaned,
    columns=cleaner.get_feature_names_out(),
    index=df.index
)
```

# **Exercise 2**
- Load **`sample_datatset.csv`**
- Replace the missings in float variables with 10 neighbours and a distance-style weight
- Replace the missings in categorical variables with the most frequent value

```python
df = pd.read_csv(gp())
```

```python
cleaner = ColumnTransformer([
    ("float", KNNImputer(n_neighbors=10, weights="distance"), make_column_selector(dtype_include="float64")),
    ("category", SimpleImputer(strategy="most_frequent"), make_column_selector(dtype_include=["object", "category"]))
])
```

```python
cleaner.fit_transform(df)
```
