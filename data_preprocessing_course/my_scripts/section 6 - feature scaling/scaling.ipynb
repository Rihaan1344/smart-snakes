```python
import pandas as pd
from get_path import get_path as gp
```

```python
df = pd.read_csv(gp())
```

```python
x = df.iloc[:, :3].dropna()
```

```python
x.head()
```

# **Normalization**
### *Also known as 0-1 scaling*

```python
from sklearn.preprocessing import MinMaxScaler
```

### **Values of a feature are scaled into a new feature, where the new minimum and maximum value are 0 and 1.**

```python
scaler = MinMaxScaler()
```

```python
(x_scaled := scaler.fit_transform(x))
```

```python
#checking max calue in each column

import numpy as np

np.apply_over_axes(np.max, x_scaled, axes=[0]) #function to apply, array to be applied on, axes to be applies on(0 => column, 1=> row)
```

```python
np.apply_over_axes(np.min, x_scaled, axes=[0]) 
```

# **Standardization**
### Subtract mean and divide by standard deviation, to get "no. of standard deviations away from mean", => 0 mean, 1 std

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

(x_scaled := scaler.fit_transform(x))
```

```python
np.apply_over_axes(np.mean, x_scaled, axes=[0])
```

```python
np.apply_over_axes(np.var, x_scaled, axes=[0])
```

# **Robust Scaling**
### Subtract median(Resistant to outliers, hence, robust), and divide by IQR -> Inter quantile range, i.e., 75% percentile - 25% percentile

```python
from sklearn.preprocessing import RobustScaler
```

```python
scaler = RobustScaler()

(x_scaled := scaler.fit_transform(x))
```

```python
np.apply_over_axes(np.median, x_scaled, axes=[0])
```

### ***Note***: As these are linear transformations, unscaled datasets can be retrieved by using invers transformation methods which are built into the scaler objects. Simply use the scaler.inverse_transform(dataset) method.
