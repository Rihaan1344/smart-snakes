```python
import pandas as pd
from get_path import get_path as gp
```

```python
from sklearn.impute import KNNImputer, SimpleImputer
```

```python
df = pd.read_csv(gp())
df.head()
```

```python
#create var x which is all rows but first three columns
x = df.iloc[:, 0:3]

x
```

# ***KNN-Imputer***

```python
knn_cleaner = KNNImputer(n_neighbors = 5)
```

```python
k_transform = knn_cleaner.fit_transform(x)
```

```python
k_transform
```

# ***Difference between mean and KNN imputers***

```python
mean_cleaner = SimpleImputer()
```

```python
m_transform = mean_cleaner.fit_transform(x)
```

```python
diff = ((k_transform - m_transform) / k_transform) * 100 #computing difference % between mean and knn
```

```python
diff
```

```python
diff.mean() #very low diff for this case
```
