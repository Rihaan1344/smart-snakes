```python
import pandas as pd
from get_path import get_path as gp
```

```python
from sklearn.feature_selection import SelectKBest, mutual_info_classif, chi2
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
```

```python
df = pd.read_csv(gp(bins = True))
```

```python
df.head()
```

```python
df["radius error"] = pd.cut(df['radius error'], 10)
```

```python
x = df.iloc[:, :10]
y = df.iloc[:, -1]
```

```python
x = OrdinalEncoder().fit_transform(x)
y = LabelEncoder().fit_transform(y)
```

```python
x
```

```python
y
```

# **Mutual information**

```python
selector = SelectKBest(lambda x, y: mutual_info_classif(x, y, discrete_features = True), k = 5)
```

```python
selector.fit_transform(x, y)
```

```python
import numpy as np
df.columns[np.append(selector.get_support(), False)]
```

# **Chi-square**

```python
selector = SelectKBest(chi2, k = 5)
```

```python
selector.fit_transform(x, y)
```

```python
df.columns[np.append(selector.get_support(), False)]
```
