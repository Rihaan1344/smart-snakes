```python
import pandas as pd
from get_path import get_path as gp
```

```python
from sklearn.feature_selection import SelectFromModel
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv(gp())
```

```python
df.head()
```

```python
df.columns[:10], df.columns[-1]
```

```python
df = df.loc[:, ['mean radius', 'mean texture', 'mean perimeter', 'mean area',
        'mean smoothness', 'mean compactness', 'mean concavity',
        'mean concave points', 'mean symmetry', 'mean fractal dimension', 'target']].dropna()
```

```python
x = df.iloc[:, :10]
y = df.iloc[:, -1]
```

```python
model = RandomForestClassifier(random_state = 0)
```

# **Threshold: Mean**

```python
selector = SelectFromModel(model, threshold = "mean") #defaults to mean, not necessary but included for readability
selector.fit_transform(x, y)
```

```python
selector.estimator_.feature_importances_
```

# **Threshold: Custom**

```python
selector = SelectFromModel(model, threshold=0.02)
```

```python
selector.fit_transform(x, y)
```

```python
selector.estimator_.feature_importances_
```

# **Threshold: Median**

```python
selector = SelectFromModel(model, threshold = "median")

selector.fit_transform(x, y)
```

```python
selector.estimator_.feature_importances_
```

# **Capping no. of features via `max_features`**

```python
selector = SelectFromModel(model, threshold = "median", max_features = 3)

selector.fit_transform(x, y)
```

# **Selecting N-most important features**

```python
import numpy as np
selector = SelectFromModel(model, threshold = -np.inf, max_features = 3)
```

```python
selector.fit_transform(x, y)
```

```python
selector.estimator_.feature_importances_
```

```python
x.columns[selector.get_support()]
```
