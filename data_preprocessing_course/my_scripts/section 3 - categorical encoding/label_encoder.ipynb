# ***NOTE:***
### **This particular encoder is meant ONLY for the target variable, as it makes false assumptions about the data which does not make sense for other normal categorical data. It is fine here since it is infact what we are trying to predict.**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
target = ["A", "B", "B", "C", "A"] #IMP: MUST BE 1D. NOT COMPATIBLE WITH 2D or 3D
```

```python
enc = LabelEncoder()
```

```python
enc.fit_transform(target)
```

### **Getting original categories back with inverse function**

```python
enc.inverse_transform(enc.fit_transform(target))
```
