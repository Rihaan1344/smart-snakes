```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
x = np.array([["A"],
              ["A"],
              ["B"],
              ["C"]]) #table with only one categorical feature and four records -> A A B C
```

```python
x
```

```python
enc = OneHotEncoder()
```

```python
enc.fit_transform(x).todense() #gives sparse matrix, converting to dense
```

### Each column in this matrix is called a **"Dummy variable"**
- ### The dummy variale is  either 0 or 1, 0 indicating a false value and 1 indicating a true value
- ### The first dummy variable asks "is this 'A'?" So the dummy variable for A for the first two entries is 1, i.e., true.
- ### The second, is for 'B'. Asks the same question. Hence the first two entries show 0 for this column while the third shows 1.
- ### Same functioning but for C for the third dummy variable

# Output with dense matrix

```python
enc = OneHotEncoder(sparse_output=False)
```

```python
enc.fit_transform(x)
```

# **Removing one variable -> k-1 dummies**
### As all of the features when encoding with k dummies add up to 1, they become linearly dependent, which is bad for linear models. They require indepent features. Hence, we drop the first dummy, i.e., we use k-1 dummies.

```python
enc = OneHotEncoder(sparse_output=False, drop='first')
```

```python
enc.fit_transform(x)
```

# **Drop if binary**
### Drops the first dummy if only 2 dummies exist
### Raises error if more than 2 dummies are present.

```python
x = np.array([["A"],
              ["C"],
              ["A"],
              ["A"]]) #only 2 dummies
```

```python
enc = OneHotEncoder(sparse_output = False, drop="if_binary")
```

```python
enc.fit_transform(x)
```

```python
y = np.array([["A"],
              ["A"],
              ["B"],
              ["C"]])
```

```python
enc = OneHotEncoder(sparse_output = False, drop="if_binary")
```

```python
enc.fit_transform(y) 
```

### No dummies were dropped

# ***Handling errors***

```python
enc = OneHotEncoder(sparse_output=False, drop = "if_binary")
```

```python
enc.fit(x) # learns x, only knows about A and C
```

```python
enc.transform(y) # does not know about B, raises error
```

### We can handle this error by using the **`handle_unknown`** parameter.
### Setting it to **`ignore`** makes the encoder fill 0 on known dummies for an unknown value.

```python
enc = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
```

```python
enc.fit(x)
```

```python
enc.transform(y)
```

### The third row, containing B, has 0 for both dummies, A and C.

# Encoding extra features
### Consider that you have a ***feature X*** in ***dataset A***. This feature X only contains values ***[P, Q, R, S***]. Now suppose you fit your one hot encoder onto this particular feature X. However, you want to train it on the same feature X, but in ***dataset B***, which contains values ***[P, Q, R, S, T]***. Due to the presence of T, our encoder will raise an error.

### Thus, in such situations were we know that there are **extra dummy variables** which are **not** present in the feature we wish to **set** to, but which may be present in other features we want to **transform**, we use the **`categories=[]`** arguement.

```python
x = [["A", "X"],
     ["B", "Y"],
     ["C", "Z"]]
```

```python
enc = OneHotEncoder(sparse_output=False, categories=[["A", "B", "C", "D"], ["X", "Y", "Z"]])

#here, we are telling our encoder: 
#my dataset has two features.
#the first feature has values ["A, B, C, D"], and not all of them are guaranteed to be present.
#the second feature has values ["X, Y, Z"], and not all of them are guaranteed to be present.
```

```python
enc.fit_transform(x)
```

### A column is created for D even though it does not appear in my array.

```python
y = [["D", "Z"]]
```

```python
enc.transform(y)
```
