```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
x = [["High"], ["Low"], ["Low"], ["Medium"]]
```

```python
enc = OrdinalEncoder()
```

```python
enc.fit_transform(x)
```

### **As order has not been defined, values were ordinally encoded alphabetically.**

### However, we want:
- ### Low -> 0
- ### Medium -> 1
- ### High -> 2

```python
enc = OrdinalEncoder(categories=[["Low", "Medium", "High"]])
```

```python
enc.fit_transform(x)
```

# **More than 1 feature**

```python
x = [["High", "A"], ["Medium", "C"], ["Low", "C"], ["Medium", "C"]]
```

```python
enc = OrdinalEncoder(categories=[["Low", "Medium", "High"], ["A", "B", "C"]])
```

```python
enc.fit_transform(x)
```
