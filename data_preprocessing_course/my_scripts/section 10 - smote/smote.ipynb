```python
import pandas as pd
from get_path import get_path as gp
```

```python
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE, SMOTENC
```

```python
df = pd.read_csv(gp()).iloc[:, -10:].dropna()
```

```python
df.head()
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
x_scaled = StandardScaler().fit_transform(x)
```

```python
y.value_counts()
```

```python
resampler = SMOTE(random_state = 42)
```

```python
x_res, y_res = resampler.fit_resample(x_scaled, y)
```

```python
y_res.value_counts()
```

# **SMOTE with categoricals**

```python
df = pd.read_csv(gp())
```

```python
df = df.loc[:, ['mean radius', 'mean texture', 'mean perimeter', 'area error', 'target']].dropna()
```

```python
x = df.iloc[:, :-1]
y = df["target"]
```

```python
y.value_counts()
```

```python
resampler = SMOTENC(categorical_features=[3], random_state=30)
```

```python
x_res, y_res = resampler.fit_resample(x, y)
```

```python
y_res.value_counts()
```
