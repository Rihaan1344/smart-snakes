```python
import pandas as pd
from get_path import get_path as gp
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv(gp())
x = df[["area error"]]
x.isnull().sum()
```

# ***Cleaning using most probable value***

```python
cleaner = SimpleImputer(strategy="most_frequent")
```

```python
cleaner.fit_transform(x)
```

```python
x.value_counts()
```
