# **Binning using quantiles**    

```python
binner = KBinsDiscretizer(strategy="quantile", n_bins = 3, encode = "ordinal")
```

```python
t = binner.fit_transform(df)
```

### **Do note that the difference between `uniform` and `quantile`, is that:**
- Uniform: divides the range into n_bins, i.e., if range of values is 10 -> 110, and number of bins is 5, then the number of elements in each bin range = (110-10) / 5 = 20. This means that 10 -> 30, 1st bin, 30 -> 50, 2nd bin, etc... notice that number of elements in each bin may vary drastically.
- Quantile: divides the data into n_bins such that each bin has approximately same, m number of elements. This means that if range of values from 10 -> 110, and number of bins = 4, but our dataset only has [1, 2, 3, 4, 5, 60, 23, 110], then, each bin becomes -> [1, 2], [3, 4], [5, 23], [60, 110]. Notice that though number of elements are roughly the same, the value of the elements in that bin varies drastically
