# Exercise 1

* Load **sample_dataset.csv**
* Consider only the first 3 columns and remove all the records that have missings
* Calculate the mean value of "mean radius"
* Use a ColumnTransformer to apply the following transformations:
    * Binarize "mean radius" using a threshold equal to the mean value
    * Binning of "mean texture" with 10 uniform bins and one-hot encoded dense output
    * Binning of "mean perimeter" with 5 quantile bins and ordinal encoding output

```python
import pandas as pd
from get_path import get_path as gp
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import Binarizer, KBinsDiscretizer
```

```python
df = pd.read_csv(gp()).iloc[:, :3].dropna()
```

```python
df.head()
```

```python
(m := float(df["mean radius"].mean()))
```

```python
transformer = ColumnTransformer([
    ("mean_radius", Binarizer(threshold = m), [0]),
    ("mean_texture", KBinsDiscretizer(n_bins = 10, strategy = "uniform", encode = "onehot-dense"), [1]),
    ("mean_perimeter", KBinsDiscretizer(n_bins = 5, strategy = "quantile", encode = "ordinal", quantile_method='averaged_inverted_cdf'), [2])
])
```

```python
transformer.fit_transform(df)
```

# Exercise 2

* Load **sample_dataset.csv**
* Consider only the first 3 columns and remove all the records that have missings
* Use a ColumnTransformer to apply the following transformations:
    * Yeo-Johnson power transform of "mean radius"
    * Box-Cox transformation of "mean texture"
    * Calculate the logarithm of "mean perimeter"

```python
import pandas as pd
from get_path import get_path as gp
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import PowerTransformer, FunctionTransformer
import numpy as np
```

```python
df = pd.read_csv(gp()).iloc[:, :3].dropna()
```

```python
df.head()
```

```python
transformer = ColumnTransformer([
    ("mean_radius_transformer", PowerTransformer(method = "yeo-johnson"), ["mean radius"]),
    ("mean_texture_transformer", PowerTransformer(method = "box-cox"), ["mean texture"]),
    ("mean_perimeter_transformer", FunctionTransformer(np.log), ["mean perimeter"])
])
```

```python
transformer.fit_transform(df)
```
