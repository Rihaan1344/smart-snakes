```python
from sklearn.preprocessing import FunctionTransformer
```

```python
import pandas as pd
import numpy as np
from get_path import get_path as gp
```

```python
df = pd.read_csv(gp()).iloc[:, :5].dropna()
```

```python
df.head()
```

```python
transformer = FunctionTransformer(
    np.log
)
```

```python
transformer.fit_transform(df)
```

# **If we want a 2D array as output**

```python
transformer = FunctionTransformer(
    np.log,
    validate=True
)
```

```python
transformer.fit_transform(df)
```

# **Using complex functions via lamba**

```python
transformer = FunctionTransformer(
    lambda x: x * 2,
    validate=True
)
```

```python
transformer.fit_transform(df)
```

# **Passing a second argument via kw_args**

```python
transformer = FunctionTransformer(
    lambda x, a: x*a,
    kw_args={"a": 3}
)
```

```python
transformer.fit_transform(df)
```

### ***Note:*** **we can also define an external function and pass it as an arg.**
