# **Exercise**
- Load **diabetes** dataset
- Apply PCA and search for the optimal number of components graphically
- Apply PCA taking 80% of variance

```python
from sklearn.datasets import load_diabetes
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
x = load_diabetes()["data"]
```

```python
x
```

```python
x.shape
```

```python
pca = PCA()
```

```python
pca.fit(x)
```

```python
bar_x = range(1, len(pca.explained_variance_ratio_) + 1)
bar_y = pca.explained_variance_ratio_

plt.bar(bar_x, bar_y)
```

```python
pca.set_params(n_components = 8)
```

```python
x_red = pca.fit_transform(x)
```

```python
pca = PCA(n_components=0.8)
```

```python
pca.fit_transform(x)
```
