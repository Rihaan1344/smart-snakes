# Are there categorical variables?

```python
cat_vars = []
for col in data.columns:
    print("-" * 50)
    print(f"Column: {col}") 
    if len(data[col].unique()) > 2:
        print(f"categories: {data[col].unique()}")
        cat_vars.append(col)
    else: print("Not categorical")
print(cat_vars)
```

# Are the numerical variables continuous, discrete or binary?

```python
for col in data.columns:
    vals = data[col].unique()
    print("-" * 50)
    print("Columns Name: {}".format(col))
    try:
        if set(vals).issubset({0, 1}):
            print("Column values are binary")
        elif all([x == int(x) for x in vals]): # check if value is same even after coercing into int
            print("Columns values are discrete")
            print("Values are: {}".format(vals))
        elif all([x != int(x) for x in vals]): # check if value is same even after coercing into int
            print("Columns values are continuous")
            print("Values are: {}".format(vals))
    except ValueError:
        print("Value is not numerical")
        
```
