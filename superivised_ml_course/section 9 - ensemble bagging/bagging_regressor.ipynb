```python
from sklearn.ensemble import BaggingClassifier, BaggingRegressor
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.preprocessing import StandardScaler

from sklearn.datasets import load_iris, load_diabetes
```

# **Classification**

```python
x, y = load_iris(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y,
    test_size=0.4,
    random_state=11
)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x_train)
```

```python
model = BaggingClassifier(LogisticRegression(),
                          n_estimators = 200, # number of learners to create
                          max_features = 2, # max number of features each learner receives
                          max_samples = 0.8, # maximum number of samples each learner receives
                          bootstrap = True, #with or without replacement of rows
                          bootstrap_features = True #with or without replacement of features
                          )
```

```python
model.fit(scaler.transform(x_train), y_train)
```

```python
model.predict(scaler.transform(x_test))
```

```python
model.score(x_test, y_test)
```

```python
model.predict_proba(scaler.transform(x_test))
```

# **Regression**

```python
x, y = load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y,
    test_size=0.4,
    random_state=11
)
```

```python
model = BaggingRegressor(LinearRegression(),
                          n_estimators = 40, # number of learners to create
                          max_features = 0.6, # max number of features each learner receives
                          max_samples = 1.0, # maximum number of samples each learner receives
                          bootstrap = True, #with or without replacement of rows
                          bootstrap_features = False #with or without replacement of features
                          )
```

```python
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test, y_test)
```
