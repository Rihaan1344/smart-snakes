```python
from sklearn.naive_bayes import CategoricalNB
```

```python
from sklearn.datasets import load_iris
```

```python
from sklearn.preprocessing import KBinsDiscretizer #numerical -> categorical
```

```python
x, y = load_iris(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y,
    test_size=0.4,
    random_state=11
)
```

```python
transformer = KBinsDiscretizer(strategy="uniform", n_bins=5, encode="ordinal")
```

```python
transformer.fit(x_train)
```

```python
x_test_categ = transformer.transform(x_test)
x_train_categ = transformer.transform(x_train)
```

```python
x_test_categ
```

```python
model = CategoricalNB()
```

```python
model.fit(x_train_categ, y_train)
```

```python
model.predict(x_test_categ)
```

```python
model.predict_proba(x_test_categ)
```
