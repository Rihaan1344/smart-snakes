```python
from sklearn.datasets import load_breast_cancer, load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

# **Binary Classification**

```python
x, y = load_breast_cancer(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y,
    test_size = 0.4,
    random_state = 11
)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x_train)
```

```python
model = LogisticRegression()
```

```python
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
model.fit(x_train_scaled, y_train)
```

```python
model.predict(x_train_scaled)
```

```python
model.predict(scaler.transform(x_test))
```

```python
model.predict_proba(scaler.transform(x_test))
```

```python
model.classes_
```

```python
model.coef_
```

# **Multiclass**

```python
x, y = load_iris(return_X_y=True)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y,
    test_size = 0.4,
    random_state = 11
)
```

```python
scaler = StandardScaler()

scaler.fit(x_train)
```

```python
model_multi = LogisticRegression()
```

```python
model_multi.fit(scaler.transform(x_train), y_train)
```

```python
model_multi.predict(scaler.transform(x_test))
```

```python
model_multi.predict_proba(scaler.transform(x_test))
```

```python
model_multi.classes_
```

```python
model_multi.coef_
```
